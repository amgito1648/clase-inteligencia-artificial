```python
import pandas as pd

# Dataset de ejemplo
data = pd.DataFrame({
    "Precio": [100, 200, 150, 300],
    "Cantidad": [2, 5, 3, 8]
})

# Crear una nueva característica: Total
data["Total"] = data["Precio"] * data["Cantidad"]
print("Datos con nueva característica:\n", data)
```

```python
# Agregar fechas al dataset
data["Fecha"] = pd.to_datetime(["2023-01-01", "2023-01-15", "2023-02-10", "2023-03-01"])


# Extraer características temporales
data["Año"] = data["Fecha"].dt.year
data["Mes"] = data["Fecha"].dt.month
data["Día"] = data["Fecha"].dt.day
data["Día_Semana"] = data["Fecha"].dt.day_name()
print("Datos con características temporales:\n", data)


```

```python
# prompt: crear un dataframe con datos sinteticos, que tenga la edad que va de 17 a 22 años, la ubicacion(latitud y longitud) en la periferia de bucaramanga, el nombre de el estudiante, la nota de inteligencia artifical de  1 a 5, dame 100 datos

import pandas as pd
import numpy as np

# Número de datos sintéticos
n_samples = 100

# Generar edades entre 17 y 22 años
edades = np.random.randint(17, 23, n_samples)

# Latitud y longitud aproximadas de la periferia de Bucaramanga
# (Ajusta estos valores según la precisión que necesites)
latitud_media = 7.12
longitud_media = -73.12
desviacion_latitud = 0.1
desviacion_longitud = 0.1

latitudes = np.random.normal(loc=latitud_media, scale=desviacion_latitud, size=n_samples)
longitudes = np.random.normal(loc=longitud_media, scale=desviacion_longitud, size=n_samples)


# Generar nombres de estudiantes aleatorios (puedes mejorar esto)
nombres = [f"Estudiante_{i}" for i in range(n_samples)]

# Generar notas de inteligencia artificial entre 1 y 5
notas_ia = np.random.randint(1, 6, n_samples)

# Crear el DataFrame
data = pd.DataFrame({
    "Edad": edades,
    "Latitud": latitudes,
    "Longitud": longitudes,
    "Nombre": nombres,
    "Nota_IA": notas_ia
})

data
```

```python
# prompt: de acuerdo a la latitud y la longitud, crear una columna que me indique la distancia a este punto llamado unab(7.1171, -73.10501) gracias

import pandas as pd
import numpy as np
from geopy.distance import geodesic

# Definir las coordenadas del punto de referencia "unab"
unab_coords = (7.1171, -73.10501)

# Función para calcular la distancia
def calcular_distancia(row):
    coords = (row["Latitud"], row["Longitud"])
    return geodesic(coords, unab_coords).kilometers

# Aplicar la función a cada fila del DataFrame para crear la nueva columna
data["Distancia_a_UNAB"] = data.apply(calcular_distancia, axis=1)

data
```
